```python
# Short Exercise 1. 

# Fozzie Bear! (based on a classic coding challenge)
# Make a program that spits out numbers and the words Fozzie Bear. 
# If the line is a multiple of 3 print Fozzie
# If the line is a multiple of 5 print Bear
# If the line is a multiple of both, print "Fozzie Bear" (with the space in between). 
# Have the program run in the range 0 to 30. 

# Answer below here: 
for i in range(30): # (yes, it was unclear whether to go from 0 to 30 inclusive or not)
    if i%3==0 and i%5==0: # check for 3 and 5
        print('Fozzie Bear')
    elif i%3==0: # check for 3 
        print('Fozzie')
    elif i%5==0: # check for 5
        print('Bear')
    else: # if no conditions satisfied, print the number
        print(i)


print()

# Alternative more general solution, which is easily scalable to any number of mod conditions.
fb = {3:'Fozzie', 5:'Bear'} 
for i in range(30): # 
    toprint=[]
    for k, v in fb.items(): # iterates through mod conditions dictionary for each number
        if i%k==0:
            toprint.append(v) # appends if condition satisfied
    print(str(i)*(not bool(len(toprint))) + ' '.join(toprint)) # first part is a boolean statement which prints the number if 'toprint' is empty, second part joins the strings in toprint with spaces

    
```

```python
#####################################
# Pattern 3. Catching bad input with 'try / except'

while True:
    a = input("Please print a number greater than 4: ")
    try: 
        if float(a) > 4: 
            print("you selected %s which is greater than 4" % a)
            break
        else:
            print("That's not greater than 4. Try again.")
    except: 
        print("I'm sorry, that was not valid input. Please enter an number")
```
