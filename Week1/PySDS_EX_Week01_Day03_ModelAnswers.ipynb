```python
# Group the files in this directory by type.
# Get all the files, 
# create a dictionary with the file extension as the key 
#   and the list of files in that directory as the value
# Print the keys and the length of the list of files so it would 
# look like the following: (the numbers will be different, of course)
#
# File type counts: 
# .py: 3
# .ipynb: 4
# .txt: 10
#
# Have two file paths. First do this with your current working directory.
# Second do this with your downloads folder. 

# Answer
import os, glob
CWD = os.getcwd()
DOWNLOADS = '~/Downloads' #### This is obviously dependent on your filesystem

allfiles= glob.glob(CWD+'/*') # get all files in directory
filetypes = set() # use a set to store the unique file types
for i in allfiles:
    if '.' in i: # identify files with an extension and add it to the filetypes set
        filetypes.add(i.split('.')[-1])

filedict = {}
for i in filetypes: # for each of the file types
    filedict[i] = glob.glob("%s%s*%s" % (CWD, os.sep, i)) # use glob to get all files with the extension

for k, v in filedict.items(): # print the summary
    print('.%s: %d' %(k, len(v)))
    
    
# and now the same for downloads - could have just put this in another for loop
allfiles= glob.glob(DOWNLOADS+'/*')
filetypes = set()
for i in allfiles:
    if '.' in i:
        filetypes.add(i.split('.')[-1])

filedict = {}
for i in filetypes: # for each of the file types
    filedict[i] = glob.glob("%s%s*%s" % (DOWNLOADS, os.sep, i))  # use glob to get all files with the extension

for k, v in filedict.items(): # print the summary
    print('.%s: %d' %(k, len(v)))
    
# Reviewer comments below here 
'''



'''
```

```python
# Stitching together 4 files. 
# The zip file: muppet_episodes_by_season.zip contains four text files
# muppet_show_season_xx.txt
# You have to merge these files together to create a single csv file.
# You have to unzip the files to a directory, read each one in
# delete the first line and append the others to a single file. 


# Answer
filepath='muppet_episodes_by_season.zip/' # edit filepath

tot_lines=[]
for i in range(1,5):
    filein = open(filepath+'muppet_show_season_%d.txt' %i, 'r') # read each file
    filelines = filein.readlines()[1:] # get the file lines, excluding line 0
    filein.close()
    tot_lines.extend(filelines) # store the file lines in larger list
         
                  
fileout = open(filepath+'merged_muppets.txt', 'w') # write to file 
fileout.writelines(tot_lines)
fileout.close()
    
fileout = open(filepath+'merged_muppets.txt', 'r') # read the file
print(fileout.read())
fileout.close()

# Reviewer comments below here 
'''


'''
```
